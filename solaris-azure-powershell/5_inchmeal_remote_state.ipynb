# Store remote state

- Configure a cloud block 
- Authenticate with Terraform Cloud `terraform login`
- Migrate the state file `terraform init`
- Delete local state file `rm terraform.tfstate`
- Configure a Service Principal 

```bash 
  # Configure a cloud block

  required_version = ">= 1.1.0" # version must be no older than v1.1.0 to support `cloud` block

  cloud {
    organization = "systema-solaris"
  workspaces {
    name = "awesome-azure-powershell"
    }

````python
# Authenticate with Terraform cloud
terraform login
````

# Migrate the state file 
- reinitialise Terraform configuration. This will cause Terraform to recognise the `cloud` block configuration.
- delete existing state on the local machine

````python
# Reinitialise and migrate after completion
terraform init
rm terraform.tfstate
````

## Check or Configure a Service Principal
- create or retrieve service principal credentials 
- Update the Terraform Cloud environment variables
- Apply the changes to the Terraform configuration

````python
# if service principal does not exist
az ad sp create-for-rbac --role="Contributor" --scopes="/subscriptions/SUBSCRIPTION_ID"

# if exists, retrieve the credentials
$credentials = Get-Credential
````

# Update the Terraform Cloud Environment Variables
In Terraform Cloud Workspace, add Azure Resource Manager's credentials as sensitive environment variables:
- ARM_CLIENT_ID : Service Principal AppID
- ARM_CLIENT_SECRET : Service Principal password
- ARM_SUBSCRIPTION_ID: Azure account subscription ID 
- ARM_TENANT_ID : Service Principal tenant

````python
# Apply the changes
terraform apply
````

Terraform is now storing the state remotely in Terraform Cloud.
