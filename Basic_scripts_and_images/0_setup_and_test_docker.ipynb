# 1. Set up directory to run Terraform projects

## Requirements:

- The Terraform CLI (1.2.0+) installed.
- The AWS CLI installed.
- AWS account and associated credentials.
- Terraform Cloud account.
- Docker Desktop installed.

## Build a Docker Container 

I use a sample main.tf configuration file to initialise Terraform in the directory. It provisions a Docker container based on the nginx image, exposes port 80 internally in the container, and maps it to port 8000 externally on the host.

I test the environment by running `terraform apply` and check the port in the browser to see if it has the default Nginx welcome page.

## Release the resources

I then use `terraform destroy` as I finish testing the environment.
