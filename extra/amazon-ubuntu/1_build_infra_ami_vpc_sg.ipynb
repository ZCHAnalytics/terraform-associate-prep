# 1. Build Infrastructure 

Create a single EC2 instance on AWS with Terraform

## 1.1. Create New Configuration

- New directory for each tf configuration `mkdir <new directory>`
- Write Configuration - `code main.tf`: Terraform Block {}, Providers, Resources
         Ubuntu image 64-bit (x86) https://eu-west-2.console.aws.amazon.com/ec2/home?region=eu-west-2#AMICatalog:
- Initialize the directory `terraform init`
- Format and validate the configuration `terraform fmt`, `terraform validate`

## 1.2. Create infrastructure

`terraform apply`


## 1.3. Inspect state

Manually Managing State

## 1.4. Troubleshooting

- Check region and ensure AMI ID is region-specific  

- Check default VPC is set in the AWS account in the correct region

## 1.5. Security Groups

Add the following arguments to the aws_instance resource:
- security group ID (vpc_security_group_ids) 
- subnet ID (subnet_id)

`terraform apply`

## Release the resources

In the end, I use `terraform destroy` to release the resources.
