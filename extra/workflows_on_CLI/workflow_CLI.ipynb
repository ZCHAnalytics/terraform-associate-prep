# The CLI-driven workflow 

The CLI-driven workflow allows to quickly iterate the configuration and work locally. The setup is as follows:

- Start the login flow `terraform login`
- Interact with Terraform API token generator in a web browser
- Add the token to the CLI
- Create a credentials variable set in Cloud/organization/settings/variable set/create variable set

Add AWS credentials as environment and sensitive and save the new variable set.

## Configuration files

I use the following configuration files:
- main.tf
- variables.tf 
- outputs.tf 
- terraform.tf 

## Create a Workspace

I update the `terraform.tf` with the names of Terraform Cloud's organisation and workspace which is 'awesome-world`.
- Run `terraform init` to create a workspace 

## Assign variable set to workspace

This workspace already has a set of global variables assigned to the organisation.


## Create infrastructure
Configure Terraform variable for instance_type

## Apply planned changes

Provision the infrastructure with `terraform apply`. Terraform Cloud references the workspace's variables and variable sets to authenticate with the AWS provider and configure an EC2 instance's name and instance type.

I can check my awesome instance in the AWS EC3 dashboard now.

## Change Infrastructure#

### Edit variables 

I can pass per-run variables to Terraform Cloud using command line flags. 

When using the CLI-driven workflow for Terraform Cloud, any variables passed using the `-var` flag will override workspace-specific variables. These new values are not saved in the workspace unless updated in the UI.

I try this ` terraform apply -var="instance_type=t2.small"` to change the EC2 instance type and the Terraform Workspace executed the run with a new value.
